```julia
using Images, TestImages, Plots
```

```julia
n = 100
H = [1/(k+j-1) for k=1:n, j=1:n]
heatmap(H)
```

```julia
plot(svdvals(H); yscale=:log10)
```

```julia
U,σ,V = svd(H)
norm(U * Diagonal(σ) * V' - H)
```

```julia

U,σ,V = svd(H)

k = 10
H_k = U[:,1:k] * Diagonal(σ[1:k]) * V[:,1:k]'
norm(H_k - H) # approx H by rank k matrix H_k
```
